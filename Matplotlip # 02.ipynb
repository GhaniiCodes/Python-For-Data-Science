```python
# importing Libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
Data = pd.read_csv("Airline Dataset.csv")
```

```python
Data.head()
```

```python
Data.tail()
```

```python
Data.info()
```

# 01

```python
Age = Data.groupby("Age")["Passenger ID"].count()
Age
```

```python
plt.hist(Data["Age"] , edgecolor="black" , bins=[10,20,30,40,50,60,70,80])

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Passenger's Ages")

plt.show()
```

# 02

```python
Gender = Data["Gender"].value_counts()
Gender
```

```python
Gender.plot(kind="bar" , color=["Blue" , "pink"] , width=0.3)

plt.xlabel("Genders")
plt.ylabel("Frequency")
plt.title("Gender Frequency")

for i , j in enumerate(Gender):
    plt.text(i , j , str(j) , ha="center" , va="bottom" , color="Red")

plt.show()

```

# 03

```python
flight_status_counts = Data.groupby(["Airport Continent", "Flight Status"]).size().unstack()
flight_status_counts
```

```python
Bar = flight_status_counts.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="viridis")


plt.title("Flight Status by Continent")
plt.xlabel("Continent")
plt.ylabel("Number of Flights")
plt.legend(title="Flight Status")


# Add value labels to each bar
for container in Bar.containers:
    Bar.bar_label(container, label_type="center", fontsize=10, color='red')


plt.show()
```

# 04

```python
Nationalities = Data.groupby("Nationality")["Passenger ID"].count()
Sorted =  Nationalities.sort_values(ascending=False).head(5)

pd.DataFrame(Sorted)


```

```python
Sorted.plot(kind="bar")

plt.xlabel("Countries")
plt.ylabel("Count Of Passengers")
plt.title("Top 5 Nations")


plt.show()
```

# 05

```python
Airports = Data.groupby("Airport Name")["Passenger ID"].count().head().sort_values(ascending=False)

pd.DataFrame(Airports)
```

```python
Airports.plot(kind="pie" , autopct='%1.1f%%' , radius=2)


plt.legend(loc= "center right")
plt.title("Top 5 Airports")


plt.show()
```

```python
Gender_Data = Data.groupby(["Airport Continent" , "Gender"]).size().unstack()
Gender_Data
```

```python
colors = ["skyblue" if gender.lower() == "male" else "hotpink" for gender in Gender_Data.columns]

Bar = Gender_Data.plot(kind="bar", stacked=True, color=colors, width=0.8)

plt.xlabel("Continents")
plt.ylabel("Frequency")
plt.title("Gender Distribution")

for containers in Bar.containers:
    Bar.bar_label(containers, label_type="center", color="black")

plt.show()
```

```python
Data["Departure Date"] = pd.to_datetime(Data["Departure Date"], dayfirst=True, errors="coerce")

Data["Month"] = Data["Departure Date"].dt.month

```

```python
Data.head()
```

```python
Data["Departure Date"] = pd.to_datetime(Data["Departure Date"], errors="coerce")
Data["Months Names"] = Data["Departure Date"].dt.strftime("%B")
```

```python
Data.head()
```

```python
Month_Data = Data["Months Names"].value_counts()
Month_Data
```

```python
plt.figure(figsize=(20,10))
Month_Data.plot(kind="line" , color="skyblue" , marker="o" , markerfacecolor = "Red" , markersize = 20 , linewidth = 5)

plt.xlabel("Months")
plt.ylabel("Number Of Flights")
plt.title("Flight Trace")
plt.grid(True, linestyle='--')




plt.show()
```
